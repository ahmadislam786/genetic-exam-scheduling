```python

def generate_exam_schedule(start_date, end_date, time_slot_duration, friday_time_slot_duration):
    exam_schedule = []

    current_date = start_date
    while current_date <= end_date:
        if current_date.weekday() < 5:
            if current_date.weekday() == 4: 
                start_time = pd.Timestamp(f"{current_date.date()} 09:00:00")
                end_time = pd.Timestamp(f"{current_date.date()} 13:00:00")
                while start_time + pd.Timedelta(hours=time_slot_duration) <= end_time:
                    exam_schedule.append((current_date.strftime("%d-%m-%Y"), current_date.strftime("%A"),
                                          start_time.strftime("%H:%M"), (start_time + pd.Timedelta(hours=time_slot_duration)).strftime("%H:%M")))
                    start_time += pd.Timedelta(hours=time_slot_duration)

                start_time = pd.Timestamp(f"{current_date.date()} 14:00:00")
                end_time = pd.Timestamp(f"{current_date.date()} 17:00:00")
                while start_time + pd.Timedelta(hours=friday_time_slot_duration) <= end_time:
                    exam_schedule.append((current_date.strftime("%d-%m-%Y"), current_date.strftime("%A"),
                                          start_time.strftime("%H:%M"), (start_time + pd.Timedelta(hours=friday_time_slot_duration)).strftime("%H:%M")))
                    start_time += pd.Timedelta(hours=friday_time_slot_duration)
            else:
                start_time = pd.Timestamp(f"{current_date.date()} 09:00:00")
                end_time = pd.Timestamp(f"{current_date.date()} 17:00:00")
                while start_time + pd.Timedelta(hours=time_slot_duration) <= end_time:
                    exam_schedule.append((current_date.strftime("%d-%m-%Y"), current_date.strftime("%A"),
                                          start_time.strftime("%H:%M"), (start_time + pd.Timedelta(hours=time_slot_duration)).strftime("%H:%M")))
                    start_time += pd.Timedelta(hours=time_slot_duration)

        current_date += pd.Timedelta(days=1)

    return exam_schedule

start_date = pd.Timestamp("2024-05-03")
end_date = pd.Timestamp("2024-05-10")
time_slot_duration = 3 
friday_time_slot_duration = 3 

exam_schedule = generate_exam_schedule(start_date, end_date, time_slot_duration, friday_time_slot_duration)

exam_schedule_df = pd.DataFrame(exam_schedule, columns=['Date', 'Day', 'Start Time', 'End Time'])

print("Generated Exam Schedule:")
print(exam_schedule_df)
```

```python
student_data = defaultdict(list)

with open('studentCourse.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        student_name = row['Student Name']
        course_code = row['Course Code']
        student_data[student_name].append(course_code)

dict(student_data)
```

```python
population_size= 5
def assign_rooms_to_students(student_data, code):

    students = [student_name for student_name, courses in student_data.items() if len(courses) >= 3 and code in courses]
    return students
```

```python
def check_clash(students_done, students_names):
    students_done_set = set(students_done)
    students_names_set = set(students_names)
   
    return bool(students_done_set & students_names_set)
```

```python
def generate_schedule(course_codes, student_data, date_time, classes, invigilators, population_size):
    population = []

    for _ in range(population_size):
        rooms_used = []
        teacher_used = []
        courses_remaining = course_codes.copy()
        schedule = []
        slot = 0
        students_done = []
        for d in date_time:
            slot += 1
            students_done = []
            if slot == 3:
                slot = 0

            for j in range(2):
                if len(courses_remaining) == 1:
                    code = courses_remaining[0]
                    courses_remaining.remove(code)
                    students_names = assign_rooms_to_students(student_data, code)
                else:
                    while True:
                        code = random.choice(courses_remaining)
                        students_names = assign_rooms_to_students(student_data, code)
                        if len(students_names) > 0:
                            break

                    clash = check_clash(students_done, students_names)

                    if clash == True:
                        for c in range(len(courses_remaining)):
                            code = random.choice(courses_remaining)
                            students_names = assign_rooms_to_students(student_data, code)
                            if len(students_names) > 0:
                                clash = check_clash(students_done, students_names)
                                if not clash:
                                    break

                    courses_remaining.remove(code)

                total_students = len(students_names)

                no_of_rooms = math.ceil(total_students/10)
                students_done.extend(students_names)

                for r in range(no_of_rooms):
                    list1 = []

                    room = random.choice(classes)
                    while room in rooms_used:
                        room = random.choice(classes)

                    rooms_used.append(room)

                    invigilator = random.choice(invigilators)
                    while invigilator in teacher_used:
                        invigilator = random.choice(invigilators)

                    teacher_used.append(invigilator)
                    list1.append(d[0])
                    list1.append(d[1])
                    list1.append((d[2], d[3]))
                    list1.append(code)
                    list1.append(invigilator)
                    list1.append(room)
                    list1.append(total_students)
                    list1.append(students_names)
                    schedule.append(list1)
                rooms_used = []
            teacher_used = []
        population.append(schedule)

    return population

rooms = range(301, 311)

population = generate_schedule(course_codes, student_data, exam_schedule_df.values, rooms, invigilators, population_size)

print(population[0])
```

```python
def calculate_fitness(schedule):
    fitness = 0

    schedule['Time Slot'] = schedule['Time Slot'].apply(lambda x: f"{x[0]}:{x[1]}")

    friday_afternoon_exams = schedule[(schedule['Day'] == 'Friday') & (schedule['Time Slot'] == '14:00')]
    if not friday_afternoon_exams.empty:
        fitness -= 10

    for i in range(len(schedule) - 1):
        current_students = set(schedule.iloc[i]['Students'])
        next_students = set(schedule.iloc[i + 1]['Students'])
        common_students_count = len(current_students.intersection(next_students))
        fitness -= common_students_count

    for index, row in schedule.iterrows():
        if 'MG' in row['Code'] and 'CS' in row['Code']:
            mg_exam_index = row['Code'].find('MG')
            cs_exam_index = row['Code'].find('CS')
            if mg_exam_index > cs_exam_index:
                fitness -= 3

    faculty_free_counts = {'Monday': 0, 'Tuesday': 0, 'Wednesday': 0, 'Thursday': 0, 'Friday': 0}
    for index, row in schedule.iterrows():
        day = row['Day']
        if '09:00' <= row['Time Slot'] <= '11:00' or '13:00' <= row['Time Slot'] <= '15:00':
            faculty_free_counts[day] += 1
    if all(count >= len(schedule) / 2 for count in faculty_free_counts.values()):
        fitness += 8

    return fitness
```

```python
fitness_values = [calculate_fitness(schedules[i]) for i in range(len(schedules))]
```

```python

def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1) - 1)
    offspring1 = parent1[:crossover_point] + parent2[crossover_point:]
    offspring2 = parent2[:crossover_point] + parent1[crossover_point:]
    return offspring1, offspring2

def mutate(schedule):
  
    mutation_type = random.choice(["swap", "change_room", "change_invigilator"])
    
    if mutation_type == "swap":
     
        exam1, exam2 = random.sample(schedule, 1)
        exam1_time_slot = exam1[2]
        exam2_time_slot = exam2[2]
        exam1[2] = exam2_time_slot
        exam2[2] = exam1_time_slot
    elif mutation_type == "change_room":

        exam = random.choice(schedule)
        new_room = random.choice(rooms)
        exam[5] = new_room
    elif mutation_type == "change_invigilator":

        exam = random.choice(schedule)
        new_invigilator = random.choice(invigilators)
        exam[4] = new_invigilator

mutation_rate = 0.1
for schedule in new_population:
    if random.random() < mutation_rate:
        mutate(schedule)

def evolve_population(population, crossover_rate, mutation_rate):
    new_population = []
    while len(new_population) < len(population):

        parent1 = random.choice(population)
        parent2 = random.choice(population)
   
        if random.random() < crossover_rate:
            offspring1, offspring2 = crossover(parent1, parent2)
            new_population.extend([offspring1, offspring2])
        else:
            new_population.extend([parent1, parent2])

    for i in range(len(new_population)):
        if random.random() < mutation_rate:
            new_population[i] = mutate(new_population[i])

    return new_population

crossover_rate = 0.8
mutation_rate = 0.11
new_population = evolve_population(population, crossover_rate, mutation_rate)
```
